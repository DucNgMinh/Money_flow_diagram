```python
def return_in_out_df(df, date):
    df['Tháng'] = date
    df = df[['Tháng', 'Trung tâm / OU Level 4', 'Trung tâm phân bổ đi / From MPA Org unit', 'Thực hiện trong kỳ / Actual this term']]
    df['Trung tâm / OU Level 4'] = df['Trung tâm / OU Level 4'].str.split(' - ').str[0]
    df['Trung tâm phân bổ đi / From MPA Org unit'] = df['Trung tâm phân bổ đi / From MPA Org unit'].str.split(' - ').str[0]

    flow_in_df = df[['Tháng', 'Trung tâm / OU Level 4', 'Thực hiện trong kỳ / Actual this term']]
    flow_in_df.rename({'Trung tâm / OU Level 4': 'Mã đơn vị tổ chức cấp 6',
                        'Thực hiện trong kỳ / Actual this term': 'Chi phí'}, axis= 1, inplace= True)

    flow_out_df = df[['Tháng', 'Trung tâm phân bổ đi / From MPA Org unit', 'Thực hiện trong kỳ / Actual this term']]
    flow_out_df['Thực hiện trong kỳ / Actual this term'] = flow_out_df['Thực hiện trong kỳ / Actual this term'] * -1
    flow_out_df.rename({'Trung tâm phân bổ đi / From MPA Org unit': 'Mã đơn vị tổ chức cấp 6',
                        'Thực hiện trong kỳ / Actual this term': 'Chi phí'}, axis= 1, inplace= True)
    return flow_in_df, flow_out_df
```

```python
flow_9_df = pd.read_excel(FILE_PATH, sheet_name= 'PB Giai đoạn 1 T9')
flow_10_df = pd.read_excel(FILE_PATH, sheet_name= 'PB Giai đoạn 1 T10')
flow_11_df = pd.read_excel(FILE_PATH, sheet_name= 'PB Giai đoạn 1 T11')
flow_11_df.rename({'Trung tâm nhận phân bổ / OU Level 4': 'Trung tâm / OU Level 4'}, axis= 1, inplace= True)
flow_12_df = pd.read_excel(FILE_PATH, sheet_name= 'PB Giai đoạn 1 T12')
flow_12_df.rename({'Trung tâm nhận phân bổ/ OU Level 4': 'Trung tâm / OU Level 4'}, axis= 1, inplace= True) 

flow_in_9_df, flow_out_9_df = return_in_out_df(flow_9_df, '2023/09')
flow_in_10_df, flow_out_10_df = return_in_out_df(flow_10_df, '2023/10')
flow_in_11_df, flow_out_11_df = return_in_out_df(flow_11_df, '2023/11')
flow_in_12_df, flow_out_12_df = return_in_out_df(flow_12_df, '2023/12')
```

```python
def cat_ma_don_vi(x):
    if len(x) == 12:
        return x[6:9]
    else:
        return x[-5:-2]
    
def cat_ma_phong(x):
    if len(x) == 12:
        return x[9:]
    else:
        return x[-2:]
```
