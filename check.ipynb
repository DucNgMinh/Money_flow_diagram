```python
df['Khối/level 2 CODE'] = df['Khối/level 2'].str.split(' - ', expand=True)[0]
df['Khối/level 2'] = df['Khối/level 2'].str.split(' - ', expand=True)[1]

df['Ban/Level 3 CODE'] = df['Ban/Level 3'].str.split(' - ', expand=True)[0]
df['Ban/Level 3'] = df['Ban/Level 3'].str.split(' - ', expand=True)[1]

df['Trung tâm/level 4 CODE'] = df['Trung tâm/level 4'].str.split(' - ', expand=True)[0]
df['Trung tâm/level 4'] = df['Trung tâm/level 4'].str.split(' - ', expand=True)[1]

df['Trung tâm nhận phân bổ / To MPA OU CODE'] = df['Trung tâm nhận phân bổ / To MPA OU'].str.split(' - ', expand=True)[0]
df['Trung tâm nhận phân bổ / To MPA OU'] = df['Trung tâm nhận phân bổ / To MPA OU'].str.split(' - ', expand=True)[1]

df
```

```python
def calculate_percentage_each_TT(tt_code_list):
    result_df = pd.DataFrame()
    for tt_code in tt_code_list:
        sub_df = df[df['Trung tâm/level 4 CODE'] == tt_code]
        sub_df['Percentage_TT'] = sub_df['Thực hiện trong kỳ / Actual in the reporting term'] / sub_df['Thực hiện trong kỳ / Actual in the reporting term'].sum() * 100
        result_df = pd.concat([result_df, sub_df])
    return result_df

result_df = calculate_percentage_each_TT(df['Trung tâm/level 4 CODE'].unique())
```
