```python
def calculate_percentage_each_TT(tt_code_list):
    result_df = pd.DataFrame()
    for tt_code in tt_code_list:
        sub_df = df[df['Trung tâm/level 4 CODE'] == tt_code]
        sub_df['Percentage_TT'] = sub_df['Thực hiện trong kỳ / Actual in the reporting term'] / sub_df['Thực hiện trong kỳ / Actual in the reporting term'].sum() * 100
        result_df = pd.concat([result_df, sub_df])
    return result_df

result_df = calculate_percentage_each_TT(df['Trung tâm/level 4 CODE'].unique())
```
